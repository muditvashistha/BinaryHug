```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url='https://www.timeout.com/music/50-best-sad-songs'
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
all_h3=soup.find_all('div',class_="_title_yvg82_9")
```

```python
song_divs = soup.find_all('div', class_="h3_19kvp_1")
songs = []

for div in song_divs:
    song_title = div.text.strip()
    songs.append(song_title)

print(f"Found {len(songs)} songs:")
for song in songs:
    print(song)
```
