```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://editorial.rottentomatoes.com/guide/great-feel-good-movies/"
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
movie_list=[]
```

```python
all_h2s = soup.find_all('h2')

# Initialize an empty list to store movie titles
movie_titles = []
movee=[]

# Loop through each h2 and extract the anchor tag text
for h2 in all_h2s:
    link = h2.find('a')
    if link:  # Check if an anchor tag exists in this h2
        dic={
            "title":link.text.strip()
        }
        movee.append(dic)
        movie_titles.append(link.text.strip())

# Print all movie titles
for movie in movie_titles:
    print(movie)
```

```python
df=pd.DataFrame(movee)
```

```python
df
```

```python
df.to_csv('Happy.csv',index=False)
```
