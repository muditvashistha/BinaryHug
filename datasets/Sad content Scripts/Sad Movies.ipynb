```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.buzzfeed.com/bendzialdowski/films-for-when-youre-sad"
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
movie_list=[]
```

```python
for movie in soup.find_all(class_='js-subbuzz__title-text'):
    print(movie.text)
    dic={
        "title":movie.text,
    }
    movie_list.append(dic)
    
```

```python
df=pd.DataFrame(movie_list)
```

```python
df
```

```python
df=df.drop([0,1]).reset_index(drop=True)
```

```python
df
```

```python
df=df.drop([34]).reset_index(drop=True)
```

```python
df
```

```python
df.to_csv('Sad.csv',index=False)
```
